```python
import os

folder_path = '/Users/wera/Max_astro/Slovakia/elisa_on_a_server/ML-EclipsingBinaries/Syntetic LC/detached/'

# Find all files with 'detached_spots_period(1-2)' in the name
files = [f for f in os.listdir(folder_path) if 'detached_spots_period(1-2)' in f]

for file in files:
    file_path = os.path.join(folder_path, file)
    
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    # Remove the first column
    df = df.iloc[:, 1:]
    
    # Write back to the same file
    df.to_csv(file_path, index=False)
```
